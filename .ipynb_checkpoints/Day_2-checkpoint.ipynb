# Exercise 6: Display numbers divisible by 5
Write a Python code to display numbers from a list divisible by 5

Expected Output:

Given list is  [10, 20, 33, 46, 55]

Divisible by 5

10

20

55

```python
list = [10,20,33,46,55]
print ("Divisible by 5")
for i in list:
    if i %5 ==0:
        print(i)
   
```

# Exercise 7: Find the number of occurrences of a substring in a string
Write a Python code to find how often the substring “Emma” appears in the given string.

Given:

str_x = "Emma is good developer. Emma is a writer"

Expected Output:

Emma appeared 2 times

```python
str_x = "Emma is good developer. Emma is a writer"
x = str_x.count("Emma")
print(f"Emma appeared {x} times.")
```

# Exercise 9: Check Palindrome Number
Write a Python code to check if the given number is a palindrome. A palindrome number reads the same forwards and backward. For example, 545 is a palindrome number.

Expected Output:

original number 121

Yes. given number is palindrome number

original number 125

No. given number is not palindrome number

```python
def palindrome(number):
    print(f"Original number",number)
    if number == int(str(number)[::-1]):
        print("Yes. given number is palindrome number")
    else:
        print("No. given number is not palindrome number")

palindrome(121)
palindrome(125)
```

# Exercise 10: Merge two lists using the following condition
Given two lists of numbers, write Python code to create a new list containing odd numbers from the first list and even numbers from the second list.

Given:

list1 = [10, 20, 25, 30, 35]

list2 = [40, 45, 60, 75, 90]

Expected Output:

result list: [25, 35, 40, 60, 90]

```python
def merge_list(list1,list2):
    result_list = []
    for num in list1:
        if num % 2 ==1:
            result_list.append(num)
    for num in list2:
        if num % 2 ==0:
            result_list.append(num)
    print(f"result list:",result_list)
    
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]    
merge_list(list1,list2)
```
